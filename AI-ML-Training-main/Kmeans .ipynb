```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data={
    'CustomeID':[1,2,3,4,5,6,7,8],
    'Annualincome':[15,16,17,28,30,45,55,60],
    'SpendingScore':[35,31,8,77,40,50,40,30]
}
df=pd.DataFrame(data)
```

```python
df
```

```python
x=df[['Annualincome','SpendingScore']]
```

```python
plt.figure(figsize=(9,7))
plt.scatter(x['Annualincome'],x['SpendingScore'],s=100,c='blue')
plt.title("custormer distribution")
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.grid(True)
plt.show()
```

```python
kmeans =KMeans(n_clusters=2,random_state=0)
df['Cluster']=kmeans.fit_predict(x)
```

```python
plt.figure(figsize=(7,8))
sns.scatterplot(data=df,x = 'Annualincome',y = 'SpendingScore',hue='Cluster',palette='Set1',s=100)
centroids=kmeans.cluster_centers_
plt.scatter(centroids[:,0],centroids[:,1],s=100,c='blue',label='Centroids')
plt.title('customer Segments :(X means)')
plt.legend()
plt.grid(True)
plt.show()
```
